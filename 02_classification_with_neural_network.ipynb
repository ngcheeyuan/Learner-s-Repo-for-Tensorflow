```python
#plot decision boundary

import numpy as np

def plot_decision_boundary(model,X,y):
  '''
  CS31n Convoluntional neural networks for visual recognition
  plots the decision boundary created by a model predicting on X
  '''

  x_min,x_max = X[:,0].min()-0.1 , X[:,0].max()+0.1
  y_min,y_max = X[:,1].min()-0.1 , X[:,1].max()+0.1

  xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
                      np.linspace(y_min,y_max,100))
  
  # create X values we're going to make predictions on these

  x_in = np.c_[xx.ravel(),yy.ravel()]

  y_pred = model.predict(x_in)
  
  if len(y_pred[0])>1:
    print('doing multiclass classification')
    y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
  
  else:
    print('doing binary classifcation')
    y_pred = np.round(y_pred).reshape(xx.shape)

  plt.figure(figsize = (16,9))
  plt.contourf(xx,yy,y_pred , cmap = plt.cm.RdYlBu,alpha =0.7)
  plt.scatter(X[:,0],X[:,1],c=y,s=40, cmap = plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

```python
from sklearn.metrics import confusion_matrix
import itertools
#make_predictions


def plot_confusion_matrix(y_true,y_pred,text_size,classes = None, figsize = (10,10)):


  cm = confusion_matrix(y_true,y_pred)
  cm_norm = cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
  n_classes = cm.shape[0]

  fig,ax = plt.subplots(figsize = figsize)
  cax = ax.matshow(cm,cmap = plt.cm.Blues)
  fig.colorbar(cax)

  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  ax.set(title = 'confusion matrix',
        xlabel = 'Predicted label',
        ylabel = 'True label',
        xticks = np.arange(n_classes),
        yticks = np.arange(n_classes),
        xticklabels = labels,
        yticklabels = labels)
  threshold = (cm.max()+cm.min())/2

  ax.xaxis.set_label_position('bottom')
  ax.xaxis.tick_bottom()

  ax.xaxis.label.set_size(text_size)
  ax.yaxis.label.set_size(text_size)
  ax.title.set_size(text_size)

  for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
    plt.text(j,i,f'{cm[i,j]} ({cm_norm[i,j]*100:.1f}%)',
            horizontalalignment = 'center',
            color = 'white' if cm[i,j] > threshold else 'black',
            size = text_size)
```

When working with images and other forms of visual data, it's a good idea to visualize as much as possible to develop a further understanding of the data and the inputs and outputs of your models.

How about we create a fun little function for:

*Plot a random image

*Make a prediction on said image

*Label the plot with the truth label &
the predicted label

```python
import random
def plot_random_image(model,images, true_labels,classes):
  '''
  Pick a random image and plot it and labels it with a prediction and truth label
  '''
  plt.figure(figsize = (10,9))
  for i in range(4):
    ax = plt.subplot(2,2,i+1)
    random_index = random.randint(0,len(images))
    target_image = images[random_index]
    prediction = model.predict(target_image.reshape(1,28,28))
    pred_label = classes[prediction.argmax()]
    true_label = classes[true_labels[random_index]]

    plt.imshow(target_image , cmap = plt.cm.binary)

    #change the color of the titles depending on if the prediction is right or wrong

    if pred_label == true_label:
      color = 'green'
    else:
      color = 'red'

    plt.xlabel(f'Pred:{pred_label} {100*tf.reduce_max(prediction):2.0f}% (True:{true_label})' , color = color)
```
